### TPC4: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = (nlugares, Vendidos, filme)
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...

listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )

...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.



````python
sala1=(150,[120],'Twilight')
sala2=(200,[131,132],'Thor: Love and Thunder')
sala3=(180,[],'''Harry Potter and the Philosopher's Stone''')

cinema=[sala1,sala2,sala3]

def listar(cinema):
    filmess=[]
    for i in cinema:
        lugares,vendidos,filmes=i
        filmess.append(filmes)
    
    return filmess


def disponivel(cinema,filme,lugar):
    available=True
    lugaresAvailable=[]
    for i in cinema: 
        lugares,vendidos,filmes=i
        lugaresAvailable.append(vendidos)

    Lista = [n for vendidos in lugaresAvailable for n in vendidos]
    
    for i in Lista:
        if lugar in Lista:
            available=False
    
    return available
    

def vendeBilhete(cinema,filme,lugar):
    di=disponivel(cinema,filme,lugar)
    if di==True:
        for i in cinema:
            lugares,vendidos,filmes=i
            if filme==filmes and lugar<=lugares:
                vendidos.append(lugar)
                vendidos.sort
                print('Os lugares ocupados são os seguintes: ',vendidos)
    else:
        print('O lugar introduzido não é válido. O número máximo de lugares para o filme ',filme,'é de',lugares)
   
    return vendidos


def inserirSala(cinema,sala): 
    if sala in cinema:
        print("A sala que tentou introduzir já se encontra no sistema!")
    else:
        cinema.append(sala)
        print("A sala foi adicionada com sucesso.")
    
    return cinema

def listardisponibilidades(cinema):
    for i in cinema:
        lugares,vendidos,filmes=i
        print("A sala do filme ",filmes, " tem ", lugares-len(vendidos)," lugares disponiveis")
    return cinema

menu2=int(input('''Pretende iniciar a aplicação? 
Se sim, selecione [1]. Se não, selecione [0]. '''))

while menu2!=0:

    menu=int(input('''Bem Vindo ao CINEMAX. Introduza o comando que pretende:

(1) Listar os filmes em exibição.
(2) Verificar se o lugar pretendido está disponível.
(3) Comprar bilhete e visualizar todos os lugares ocupados até ao momento.
(4) Visualizar os filmes disponíveis e correspondente ocupação.
(5) Inserir uma nova sugestão de filme.
(6) Encerrar o programa.'''))

    
    if menu==1:
        print('Os filmes disponíveis são os seguintes: ', listar(cinema))
   
    elif menu==2:
        opc1=input('Filme que pretende ver: ')
        opc2=int(input('Lugar que pretende ficar: '))
        disponivel(cinema,opc1,opc2)

    if menu==3:
        opc3=input('Introduza o filme que quer ver: ')
        opc4=int(input('Introduza o lugar que pretende escolher: '))
        vendeBilhete(cinema,opc3,opc4)
    
    
    if menu==4:
        listardisponibilidades(cinema)
       
    if menu==5:
        opc5=input('Filme que quer adicionar: ')
        opc6=int(input('Número de lugares disponíveis na sala a ser adicionada: '))
        inserirSala(cinema,opc5,opc6)


else:
    print('O programa irá encerrar. Obrigada pela sua visita!')
````
