### TPC6: Processando um dataset em CSV

Escreve um programa que lê o dataset `obras.csv` e realiza as seguintes tarefas (tenta isolar cada tarefa numa função):
* Carrega o dataset para uma estrutura de dados pensada por ti;
* Quantas obras existem catalogadas no dataset?
* Imprime no monitor uma tabela com o título da obra, a sua descrição, o seu compositor e ano de criação;

* Produz uma lista de tuplos `(título, ano)` ordenada crescentemente por ano;
* Produz uma lista ordenada dos compositores;
* Calcula uma distribuição das obras por período;
* Calcula uma distribuição das obras por ano;
* Calcula uma distribuição das obras por compositor;
* Especifica uma função que recebendo uma distribuição desenha o seu gráfico. Aplica-a às distribuições anteriores;
* Problema da inversão estrutural: calcula uma estrutura de dados que corresponde a uma lista dos compositores em que cada compositor tem a ele associado uma lista dos títulos das obras que compôs;
* Cria uma função de visualização para estrutura de dados calculada na alínea anterior.


```python
import TP6

obras = TP6.leObras("obras.csv")

def menu(): 

    opcao = input ('''
        Introduza o que que pretende ver: 
                          (1) Para ver quantas obras existem catalogadas no dataset; 
                          (2) Para imprimir no monitor uma tabela com o título da obra, a sua descrição, o seu compositor e ano de criação;
                          (3) Para ordenar alfabeticamente o título das obras;
                          (4) Para ordenar as obras por ano;
                          (5) Mostrar uma lista ordenada dos Compositores;
                          (6) Mostrar a distribuição das Obras por Período;
                          (7) Mostrar a distribuição das obras por Ano;
                          (8) Mostrar a distribuição das obras por Compositor;
                          (9) Gráfico da distribuição das obras por Período;
                          (10) Gráfico da distribuição das Obras por Ano;
                          (11) Gráfico da distribuição das Obras por Compositor;
                          (12) Lista dos Compositores em que cada Compositor tem a ele associado uma lista dos Títulos das Obras que compôs;
                        
                         ''')
         
    return int(opcao)   



def opcoes():
   
    opcao=25
    
    
    while opcao != 0 :
        opcao= menu()
        
        if opcao == 1:
          resultado=TP6.NObras(obras)
          print(resultado)
            
        elif opcao == 2:
            resultado= TP6.Imprime(obras)
            print(resultado)
            
        if opcao == 3:
            resultado = TP6.titleAno(obras)
            print(resultado)
        
        elif opcao == 4:
            resultado = TP6.titleAno_2(obras)
            print (resultado)
         
        if opcao == 5:
            resultado = TP6.OrdComp(obras)
            print(resultado)
        
        elif opcao == 6:
            resultado= TP6.distribPeriodo(obras)
            print(resultado)
                
        if opcao == 7:
            resultado = TP6.titPorAno(obras)
            print(resultado)
                
        elif opcao == 8:  
            resultado = TP6.titPorComp(obras)
            print(resultado)
        
        if opcao == 9: 
            resultado = TP6.Grafico1(obras)
            print(resultado)
        
        elif opcao == 10:
            grafico = TP6.Grafico2(obras)
            print(grafico)
        
        if opcao == 11:
            grafico = TP6.Grafico3(obras)
            print(grafico)
        
        elif opcao == 12:
            grafico = TP6.inversao(obras):
            print(grafico)
        
               
        
opcoes()   
```
