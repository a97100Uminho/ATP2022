### TPC3: Aplicação para manipulação de listas de inteiros
- Crie uma aplicação em Python que coloca no monitor o seguinte menu:
    * (1) Criar Lista 
    * (2) Ler Lista
    * (3) Soma
    * (4) Média
    * (5) Maior
    * (6) Menor
    * (7) estaOrdenada por ordem crescente
    * (8) estaOrdenada por ordem decrescente
    * (9) Procura um elemento
    * (0) Sair
- O utilizador irá escolher uma das opções introduzindo o número correspondente;
- Se a opção não for sair, a aplicação executa a operação pretendida, apresenta o resultado e a seguir apresenta de novo o menu;
- Se a opção for sair, a aplicação termina colocando uma mensagem no monitor.

* No desenvolvimento da aplicação deverá ter em atenção o seguinte:
    - A aplicação terá uma variável interna para guardar uma lista de números;
    - Na opção 1, deverá ser criada uma lista de números aleatórios entre 1 e 100 que será guardada na variável interna;
    - Na opção 2, deverá ser criada uma lista com números introduzidos pelo utilizador, que será guardada na variável interna;
    - *Nestas primeiras opções, se a variável interna já tiver uma lista, esta será sobreposta/apagada pela nova lista;*
    - Na opção 3, será calculada a soma dos elementos na lista no momento;
    - Na opção 4, será calculada a média dos elementos na lista no momento;
    - Na opção 5, será calculado o maior elemento da lista no momento;
    - Na opção 6, será calculado o menor elemento da lista no momento;
    - Na opção 7, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem crescente;
    - Na opção 8, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem decrescente;
    - Na opção 9, a aplicação irá procurar um elemento na lista, se o encontrar deverá devolver a sua posição;
    - Se o utilizador selecionar a opção 0, a aplicação deverá terminar mostrando a lista que está nesse momento guardada.


```python
import random

vInterna=[]

menu2=int(input('''Pretende iniciar a aplicação? 
Se sim, selecione [1]. Se não, selecione [0]. '''))

while menu2!=0:

    menu=int(input('''Introduza o número da opção que pretende, de acordo com o seguinte:
        * (1) Criar Lista 
        * (2) Ler Lista
        * (3) Soma
        * (4) Média
        * (5) Maior
        * (6) Menor
        * (7) estaOrdenada por ordem crescente
        * (8) estaOrdenada por ordem decrescente
        * (9) Procura um elemento
        * (0) Sair'''))


    if menu==1:

        N=100
        
        vInterna.clear()
        
        i=0
        while i<101:
            vInterna.append(random.randrange(1,101))
            i=i+1

        print(vInterna)

    if menu==2:
        
        vInterna.clear()

        N=int(input("Introduza o número de valores que serão introduzidos na lista: "))
        
        def leLista(N):
            i=0
            while(i<N):
                x=int(input("Insira um número: "))
                vInterna.append(x)
                i=i+1 

            return vInterna
        
        print("A lista criada é a seguinte: ",leLista(N))

    if menu==3:        
        
        def somaLista(vInterna):

            soma=0            
            for i in vInterna:
                soma=soma+i
                    
            return soma       
        
        print("A soma dos valores presentes na lista é a seguinte: ",somaLista(vInterna))

    if menu==4:

        def media(vInterna,N):
            
            media2=somaLista(vInterna)/N

            return media2
        
        print("A média dos valores presentes na lista criada é a seguinte: ",media(vInterna,N))

    if menu==5:
        
        def maiorLista(vInterna):

            max=vInterna[0]
            for i in vInterna[1:]:
                if i>max:
                    max=i

            return max

        print("O maior valor presente na lista criada é o seguinte: ",maiorLista(vInterna))

    if menu==6:
        
        def menorLista(vInterna):

            min=vInterna[0]
            for i in vInterna[1:]:
                if i<min:
                    min=i

            return min

        print("O menor valor presente na lista criada é o seguinte: ",menorLista(vInterna))

    if menu==7:

        def crescente(vInterna):

            menor=vInterna[0]
            for elem in vInterna:
                if elem<menor:
                    menor=elem
            return menor


        listaCrescente=[]
        vInterna_modificada=vInterna.copy()
        
        while len(vInterna_modificada)!=0:
            e=crescente(vInterna_modificada)
            listaCrescente.append(e)
            vInterna_modificada.remove(e) 
        
        if vInterna==listaCrescente:
            print('Sim!')
        else:
            print("Não!")

    if menu==8:
        
        def decrescente(vInterna):

            maior=vInterna[0]
            for elem in vInterna:
                if elem>maior:
                    maior=elem
            return maior


        listaDecrescente=[]
        vInternaModificada=vInterna.copy()
        
        while len(vInternaModificada)!=0:
            e=decrescente(vInternaModificada)
            listaDecrescente.append(e)
            vInternaModificada.remove(e) 
        
        if vInterna==listaDecrescente:
            print('Sim!')
        else:
            print("Não!")


    if menu==9:

        n=int(input("Introduza o valor a ser procurado na lista: "))
        for i in vInterna: 
            if n==i:
                b=vInterna.index(n)

        print("O índice do valor", n,"é o seguinte: ",b)


print("selecionado o comando de 'exit', a aplicação irá encerrar.")
```
